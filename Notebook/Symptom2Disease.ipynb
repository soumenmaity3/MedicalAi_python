# Model

## Step 1: Setup

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.utils.data import DataLoader, Dataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import torch
import pandas as pd
import json
```

## Step 2: Prepare your dataset

```python
df = pd.read_csv("final_disease-symptom-description-dataset.csv")

label_to_id = {label: idx for idx, label in enumerate(sorted(df["label"].unique()))}
id_to_label = {v: k for k, v in label_to_id.items()}

df["label_id"] = df["label"].map(label_to_id)

# Save label mapping (VERY IMPORTANT)
with open("label_mapping.json", "w") as f:
    json.dump(label_to_id, f)
```

## Step 3: Train / validation split

```python
train_df, val_df = train_test_split(
    df,
    test_size=0.2,
    stratify=df["label_id"],
    random_state=42
)
```

## Step 4: Dataset class (no padding here)

```python
class TextDataset(Dataset):
    def __init__(self, dataframe):
        self.df = dataframe

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        return {
            "text": self.df.iloc[idx]["text"],
            "label": self.df.iloc[idx]["label_id"]
        }
```

## Step 5: Dynamic padding with collate function

```python
def collate_fn(batch):
    texts = [item["text"] for item in batch]
    labels = torch.tensor([item["label"] for item in batch], dtype=torch.long)

    encoding = tokenizer(
        texts,
        padding=True,
        truncation=True,
        max_length=128,
        return_tensors="pt"
    )

    encoding["labels"] = labels
    return encoding
```

## Step 6: Load model

```python
model_name = "dmis-lab/biobert-v1.1"
tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    num_labels=len(label_to_id)
)
```

## Step 7: DataLoaders

```python
train_loader = DataLoader(
    TextDataset(train_df),
    batch_size=16,
    shuffle=True,
    collate_fn=collate_fn
)

val_loader = DataLoader(
    TextDataset(val_df),
    batch_size=16,
    shuffle=False,
    collate_fn=collate_fn
)
```

## Step 8: Training + evaluation loop

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

optimizer = torch.optim.AdamW(model.parameters(), lr=2e-5)

for epoch in range(3):
    # ---- Training ----
    model.train()
    total_loss = 0

    for batch in train_loader:
        batch = {k: v.to(device) for k, v in batch.items()}

        outputs = model(**batch)
        loss = outputs.loss

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    avg_train_loss = total_loss / len(train_loader)

    # ---- Validation ----
    model.eval()
    preds, true_labels = [], []

    with torch.no_grad():
        for batch in val_loader:
            batch = {k: v.to(device) for k, v in batch.items()}
            outputs = model(**batch)

            logits = outputs.logits
            predictions = torch.argmax(logits, dim=1)

            preds.extend(predictions.cpu().numpy())
            true_labels.extend(batch["labels"].cpu().numpy())

    acc = accuracy_score(true_labels, preds)

    print(f"Epoch {epoch+1} | Train Loss: {avg_train_loss:.4f} | Val Acc: {acc:.4f}")
```

## Step 9: Save model

```python
import os

SAVE_DIR = os.path.abspath("biobert_symptom_classifier")
os.makedirs(SAVE_DIR, exist_ok=True)

model.save_pretrained(SAVE_DIR)
tokenizer.save_pretrained(SAVE_DIR)

print("Model saved to:", SAVE_DIR)
```

# Prediction / Inference Code (clean & reusable)

## 1. Load model, tokenizer, and label mapping

```python
import torch
import json
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch.nn.functional as F
```

```python
import os
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

MODEL_PATH = os.path.abspath("biobert_symptom_classifier")

tokenizer = AutoTokenizer.from_pretrained(MODEL_PATH)
model = AutoModelForSequenceClassification.from_pretrained(MODEL_PATH)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
model.eval()

print("Local model loaded successfully")
```

## 2. Core prediction function (THIS IS IMPORTANT)

```python
!pip install pyspellchecker
```

```python
from spellchecker import SpellChecker

spell = SpellChecker()

def correct_spelling(text):
    words = text.lower().split()
    corrected = []

    for w in words:
        if w.isalpha():
            corrected.append(spell.correction(w) or w)
        else:
            corrected.append(w)

    return " ".join(corrected)
```

```python
import torch.nn.functional as F

def predict_specialty(text):
    text = correct_spelling(text)
    inputs = tokenizer(
        text,
        truncation=True,
        padding=True,
        max_length=128,
        return_tensors="pt"
    )
    inputs = {k: v.to(device) for k, v in inputs.items()}

    with torch.no_grad():
        outputs = model(**inputs)

    probs = F.softmax(outputs.logits, dim=1)
    conf, pred_id = torch.max(probs, dim=1)

    return pred_id.item(), conf.item()
```

## 3. Test prediction (do this once)

```python
pred_id, confidence = predict_specialty(correct_spelling('chest pain'))

predicted_label = id_to_label[pred_id]

print("Predicted specialty:", predicted_label)
print("Confidence:", round(confidence, 3))
```
